```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# I wanted to set up a social network that might be able to track Covid infections among contacts and beyond. 
# I set up a dataset that was suitable for a compelling Krackhardt kite graph using NetworkX.  

```

```python
df = pd.read_csv('https://raw.githubusercontent.com/evanmclaughlin/DATA-620/main/covid_infections.v2.csv')

g = nx.Graph()
g.add_edges_from(df[['patient', 'infected_contact']].values)
```

```python
nx.draw(g, with_labels=True, font_size = 6, font_weight='bold')
plt.show()
```
